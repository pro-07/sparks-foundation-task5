## Retailers Selling Paper,Labels,Envelopes,Fasteners have got 100% profit.

```python
sns.scatterplot(x="Sales",y="Profit",data=df)
plt.show()
```

### Here it can be seen that it is not necessary that with the increase in sale profit increases.

```python
plt.figure(figsize=(12,8))
sns.barplot(x="Sub-Category",y="Profit",data=df)
plt.xticks(rotation=90)
plt.show()
```

<h3>Here we can see that the tables have more loss or less profit than any other sub category while the copiers has more profit.</h3>

```python
df['Discount'].value_counts()
```

```python
df['Discount'].unique()
```

```python
plt.scatter(x='Discount',y='Profit',data=df)
plt.show()
```

<h3>From the above graph we can see that the discount more than 0.3 gives more loss in profit while the discount rate less 0.3 has more profit.</h3>

```python
plt.scatter('Discount','Sales',data=df)
plt.show()
```

<h3>We can see that the Discount keeps increasing the sales decreases.</h3>

```python
sns.barplot(x='Category',y='Profit',data=df)
plt.title('Profit by Category wise')
plt.show()
```

<h3>From the above graph we can see that profit from Technology section are more than Furniture and Office Supplies.</h3>

```python
sns.barplot(x='Category',y='Sales',data=df)
plt.title('Sales by Category wise')
plt.show()
```

<h3>There is more sales in Technology and Furniture while the Office Supplies has less sales.</h3>

# Weak Areas

1. The discount rate should not be increased more than 0.3.
2. The Furniture category has more loss than any other category.The company should focus more on Technology and Office Supplies.
3. There is huge  profit loss from discount rate as it  keeps increasing more than 0.3
4. The Shipping charges should be increased if the delivery is on the 'Same Day'.
